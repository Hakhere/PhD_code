```python
xls = pd.ExcelFile (path/'PHERC_data_2022_Aug.xls')
df = pd.read_excel(xls, 'B8_fan')
df
```

```python
df.columns
```

```python
clean_df = df[['OPI_DATE', 'STATE_DATA']]
# Change OPI_DATE format
clean_df['OPI_DATE'] = clean_df['OPI_DATE'].apply(lambda x: datetime.strptime(x[:-7], '%Y-%m-%d %H:%M:%S'))
# Get Date 
clean_df['STATE_DATA'] = clean_df['STATE_DATA'].apply(lambda x: x[0])
clean_df
```

```python
start_time = clean_df['OPI_DATE'].min()
finish_time = clean_df['OPI_DATE'].max()
```

```python
timestamps = []
time = start_time.replace(second=0, minute=0)
while time <= finish_time + timedelta(hours=1):
    timestamps.append(time)
    time = time + timedelta(hours=1)
result_df = pd.DataFrame(timestamps, columns=['Timestamp'])
result_df['Fan Time (minute)'] = ''
result_df['Fan Time Per Hour'] = ''
result_df
```

```python
current_state = clean_df.loc[0, 'STATE_DATA']
current_row = 0
on_start = clean_df.loc[0, 'OPI_DATE']
on_stop = clean_df.loc[0, 'OPI_DATE']
for index, row in result_df.iterrows():
    
    while clean_df.loc[current_row, 'OPI_DATE'] <= row['Timestamp'] \
          and clean_df.loc[current_row, 'STATE_DATA'] == 'F' \
          and current_row + 1 < clean_df['OPI_DATE'].count():
        current_row = current_row + 1
        
    if current_state == 'F' and clean_df.loc[current_row, 'STATE_DATA'] == 'T':
        on_start = clean_df.loc[current_row, 'OPI_DATE']
        while clean_df.loc[current_row, 'STATE_DATA'] == 'T' and current_row + 1 < clean_df['STATE_DATA'].count():
            current_row = current_row + 1
        on_stop = clean_df.loc[current_row, 'OPI_DATE']
#         print(on_start, '---------', on_stop)
    
    if row['Timestamp'] >= on_start.replace(second=0, minute=0) and row['Timestamp'] <= on_stop.replace(second=59, minute=59):
        result_df.at[index, 'Fan Time (minute)'] = 60
        if row['Timestamp'].date() == on_start.date() and \
           row['Timestamp'].hour == on_start.hour:
            result_df.at[index, 'Fan Time (minute)'] = result_df.loc[index, 'Fan Time (minute)'] - on_start.minute
        
        if row['Timestamp'].date() == on_stop.date() and \
           row['Timestamp'].hour == on_stop.hour:
            result_df.at[index, 'Fan Time (minute)'] = result_df.loc[index, 'Fan Time (minute)'] - (60 - on_stop.minute)
        
        result_df.at[index, 'Fan Time Per Hour'] = result_df.loc[index, 'Fan Time (minute)'] / 60
        
    
    current_state = clean_df.loc[current_row, 'STATE_DATA']    
```
