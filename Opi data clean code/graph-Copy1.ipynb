```python
from pathlib import Path
import pandas as pd
from tqdm.notebook import tqdm 
from datetime import  datetime, timedelta
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.dates import DateFormatter, MonthLocator
```

```python
# Set the directory
# Use this code for plotting the graphs of OPI data after the inital clean up
# Ensure the excel files as CVS delimited
path = Path('U:') / 'Opi system file' / 'OPI CO2 Nov3 2020'
```

```python
# Original file name
filename = path / 'CO2 3Nov 2020 Bin13.csv'
```

```python
# Load file into Pandas data frame
df = pd.read_csv(filename)
df
```

```python
# Change the date format
for index, row in tqdm(df.iterrows()):

    # Remove unnecasy charactor
    timestamp_text = row.Date_CO2
    try:
        # Convert to Python datetime variable
        timestamp = datetime.strptime(timestamp_text, '%m/%d/%Y %H:%M')
        # Replace the old data
        df.at[index, 'Date_CO2'] = timestamp
    except:
        pass

# View new version
df
```

```python
# Change the date format
for index, row in tqdm(df.iterrows()):

    # Remove unnecasy charactor
    timestamp_text = row.Date_Ambient
    try:
        # Convert to Python datetime variable
        timestamp = datetime.strptime(timestamp_text, '%m/%d/%Y %H:%M')
        # Replace the old data
        df.at[index, 'Date_Ambient'] = timestamp
    except:
        pass

# View new version
df
```

```python
# Change the date format
for index, row in tqdm(df.iterrows()):

    # Remove unnecasy charactor
    timestamp_text = row.Date_Grain
    try:
        # Convert to Python datetime variable
        timestamp = datetime.strptime(timestamp_text, '%m/%d/%Y %H:%M')
        # Replace the old data
        df.at[index, 'Date_Grain'] = timestamp
    except:
        pass

# View new version
df
```

```python
date_grain_df = df[['Date_Grain', 'Grain_Avg_Temp', 'Grain_Avg_RH']].copy()
date_grain_df
```

```python
date_grain_df = date_grain_df.dropna()
date_grain_df
```

```python
ambient_df = df[['Date_Ambient', 'Temp_Ambient', 'RH_Ambient']].copy()
ambient_df = ambient_df.dropna()
ambient_df
```

```python
co2_df = df[['Date_CO2', 'CO2_Plenum', 'CO2_Headspace']].copy()
co2_df = co2_df.dropna()
co2_df
```

```python
fig, ax = plt.subplots(dpi=150)
ax.plot(date_grain_df.Date_Grain.to_numpy(), date_grain_df.Grain_Avg_Temp.to_numpy(), 'r--')
ax.plot(date_grain_df.Date_Grain.to_numpy(), date_grain_df.Grain_Avg_RH.to_numpy(), 'b')
```

```python
fig, ax = plt.subplots(dpi=150)
# line1 = ax.plot(co2_df.Date_CO2.to_numpy(), co2_df.CO2_Plenum.to_numpy(), color='#cc6408', label='CO2_Plenum', alpha=0.5) ## helps to dim colors
line1 = ax.plot(co2_df.Date_CO2.to_numpy(), co2_df.CO2_Plenum.to_numpy(), color='#cc6408', label='CO2_Plenum')
line2 = ax.plot(co2_df.Date_CO2.to_numpy(), co2_df.CO2_Headspace.to_numpy(), 'blue', label='CO2_Headspace')
ax2 = ax.twinx()
line3 = ax2.plot(date_grain_df.Date_Grain.to_numpy(), date_grain_df.Grain_Avg_Temp.to_numpy(), 'g', label='Grain_Avg_Temp')
# line4 = ax2.plot(date_grain_df.Date_Grain.to_numpy(), date_grain_df.Grain_Avg_RH.to_numpy(), 'brown', label='Grain_Avg_RH')
line5 = ax2.plot(ambient_df.Date_Ambient.to_numpy(), ambient_df.Temp_Ambient.to_numpy(), 'purple', label='Grain_Ambient')
date_form = DateFormatter("%m/%d/%Y")
ax.xaxis.set_major_formatter(date_form)
months = MonthLocator()
ax.xaxis.set_major_locator(months)
fig.suptitle('BIN 13', y=1.1)
ax.set_xlabel('Date')
ax.set_ylabel('$CO_2$ x 1000 ppm');
ax2.set_ylabel('Temperature, $\degree$F');
lns = line5 + line1 + line2 + line3
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, bbox_to_anchor=(0.5, 1.2), loc="upper center", ncol=3);
plt.xlim([datetime(2020, 11, 6), datetime(2021, 3, 5)])
filename = path / 'Bin 13.jpeg'
# plt.savefig(filename)
```
