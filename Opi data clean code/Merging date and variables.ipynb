```python
filename = path / 'BIN5_masterlist.xlsx'
df = pd.read_excel(filename, sheet_name = 'Sheet1')
df
```

```python
#First variable to use noting that the date here is priority
df1 = df[['OPI_DATE', 'Center_cable', 'North_cable', 'East_cable', 'South_cable', 'West_cable', 'Grain_bulk_temp']].dropna().set_index('OPI_DATE')
df1
```

```python
time1 = df1.index
time1
```

```python
#SECOND VARIABLE 
df2 = df[['Date', 'Headspace_temp', 'Headspace_RH', 'Headspace_CO2']].dropna()
df2
```

```python
time2 = df2['Date'] #SECOND VARIABLE DATE
time2
```

```python
data = []
for timestamp in time1:
    min_index = np.argmin(np.abs(time2 - timestamp))  #argmin gives the lowest index on the dataframe
    row = df2.iloc[min_index]
    data.append({
        'OPI_DATE': timestamp,
        'Center_cable': df1.loc[timestamp]['Center_cable'],
        'North_cable': df1.loc[timestamp]['North_cable'],
        'East_cable': df1.loc[timestamp]['East_cable'],
        'South_cable': df1.loc[timestamp]['South_cable'],
        'West_cable': df1.loc[timestamp]['West_cable'],
        'Grain_bulk_temp': df1.loc[timestamp]['Grain_bulk_temp'],
        'Date': row['Date'],
        'Headspace_temp': row['Headspace_temp'],
        'Headspace_RH': row['Headspace_RH'],
        'Headspace_CO2': row['Headspace_CO2'],

    })
new_df = pd.DataFrame(data)
new_df
    
```
