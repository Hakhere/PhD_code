```python
path = Path('W:') / 'Amber Ag' / 'Amber clean data'
```

```python
filename = path / 'Bin_5_simulation_start.xlsx'
df = pd.read_excel(filename)
```

```python
for index, row in df.iterrows():
    if row['Date'] is pd.NaT:
        # print(df.loc[index-1, 'Date'] + timedelta(hours=1))
        df.loc[index, 'Date'] = df.loc[index-1, 'Date'] + timedelta(hours=1)
```

```python
df['Date'] = df['Date'].apply(lambda x: datetime.strftime(x, '%m/%d/%Y %I:%M %p'))
```
