```python
from pathlib import Path
import pandas as pd
from tqdm.notebook import tqdm 
from datetime import  datetime, timedelta
import numpy as np
```

```python
# Set the directory
#Run this first for OPI cleaning
# OPI CLEANING AND AVERAGING
path = Path('U:') / 'Opi system file' / 'SQL database'/ '28Feb21'
```

```python
# Original file name
filename = path / 'PHERC RAW BIN15.xls'
```

```python
# Load file into Pandas data frame
df = pd.read_excel(filename)
df
```

```python
# Reload data
df = pd.read_excel(filename)

# Change the date format
for index, row in tqdm(df.iterrows()):

  # Remove unnecasy charactor
  timestamp_text = row.OPI_DATE[:16]
  # Convert to Python datetime variable
  timestamp = datetime.strptime(timestamp_text, '%Y-%m-%d %H:%M')
  # Convert back to desired format
  new_timestamp = datetime.strftime(timestamp, '%m/%d/%Y %I:%M %p')
  # Replace the old data
  df.at[index, 'OPI_DATE'] = new_timestamp

# View new version
df
```

```python
cols = df.columns.tolist()
cols
```

```python
# Create new dataframe with selected columns
keep_cols = ['OPI_DATE', 'BIN_MC_MAX', 'BIN_MC_MIN', 'BIN_MC_AVG']
new_df = df[keep_cols]
new_df
```

```python
avg_df = pd.DataFrame(df['OPI_DATE'], columns=['OPI_DATE'])
avg_df
```

```python
# Spread data out
for index, row in tqdm(df.iterrows()):
  for sensor in ['C', 'RH', 'MC']:
    total = []
    for i in range(5):
      col_name = f'{sensor}{i+1}'
      raw_text = row[col_name]
      raw_data = np.array(raw_text.split()[:6], dtype=np.float)
      if sensor == 'C':
        sensor_new = 'T'
        filter = raw_data > -40
      elif sensor == 'RH':
        sensor_new = sensor
        filter = raw_data > 0
      elif sensor == 'MC':
        sensor_new = sensor
        filter = raw_data > 5
      
      filtered_data = raw_data[filter]
      average = filtered_data.mean()
      avg_df.loc[index, f'{sensor_new}{i+1}-avg'] = average
      total.extend(filtered_data.tolist())
      for j, data in enumerate(raw_text.split()):
        new_col_name = f'{sensor_new}{i+1}-{j+1}'
        new_df.loc[index, new_col_name] = data
    total_avg = np.array(total).mean()
    avg_df.loc[index, f'{sensor_new}-totalavg'] = total_avg
```

```python
# Spread data out again
for index, row in tqdm(df.iterrows()):
  for sensor in ['CABLE_MAX', 'CABLE_MIN', 'CABLE_AVG']:
    raw_text = row[sensor]
    average = np.array(raw_text.split(), dtype=np.float).mean()
    avg_df.loc[index, f'{sensor}-avg'] = average
    for i, data in enumerate(raw_text.split()):
      new_col_name = f'{sensor}-{i+1}'
      new_df.loc[index, new_col_name] = data

```

```python
current_time = datetime.strptime(avg_df.at[0, 'OPI_DATE'], '%m/%d/%Y %I:%M %p')
for index, row in avg_df.iterrows():
  timestamp = datetime.strptime(row.OPI_DATE, '%m/%d/%Y %I:%M %p')
  if (timestamp - current_time) > timedelta(minutes=59) or index == 0:
    current_time = timestamp
  else:
    avg_df = avg_df.drop(index)
```
