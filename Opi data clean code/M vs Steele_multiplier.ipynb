```python
xls = pd.ExcelFile ('W:/Amber Ag/Amber clean data/Bin 13 review/Bin13_steele_cal_update.xlsx')
cf = pd.read_excel(xls, sheet_name = 'Bin13_steele_update')
cf
```

```python
cf['CO2_Headspace'] = np.nan
cf['CO2_Plenum'] = np.nan
```

```python
mf['Date_CO2'] = mf['Date_CO2'].apply(lambda x: x.replace(minute=0, second=0))
mf = mf.drop_duplicates(subset='Date_CO2')
mf = mf.set_index('Date_CO2')
mf
```

```python
index_list = mf.index.to_list()
for index, row in cf.iterrows():
    calculate_time = row['OPI_DATE'].replace(minute=0)  
    if calculate_time in index_list:
        cf.loc[index, 'CO2_Headspace'] = mf.loc[calculate_time, 'CO2_Headspace']
 
```
